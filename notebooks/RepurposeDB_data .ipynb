```python
disease_name = nodes.loc[nodes['kind']=='Disease'].values.tolist()
```

```python
ddp=[]
for _,row in data[['name','target']].iterrows():
    if row['name'].lower() in common_compound and row['target'].lower() in common_disease:
        
        if nodes.loc[nodes['name']==row['name'].lower(),'id'].item() and nodes.loc[(nodes['name']== row['target'].lower()) & (nodes['kind']=='Disease'),'id'].item():
            drug = nodes.loc[nodes['name']==row['name'].lower(),'id'].item()
            print(drug)
            disease = nodes.loc[(nodes['name']== row['target'].lower()) & (nodes['kind']=='Disease'),'id'].item()
            print(disease)
            ddp.append((drug,disease))
        else:
            continue
```

```python
len(np.unique(ddp))
```

```python
b =list(set(ddp))
```
