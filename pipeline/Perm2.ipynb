```python
import pandas as pd
dfhimnode_2 = pd.read_csv('permu_data/nodes-2.tsv',sep='\t')

dfhimedge_2 = pd.read_csv('permu_data/edges-2.sif.gz',sep='\t')

himgraph_2 = nx.Graph()
for index,row in dfhimnode_2.iterrows():
    if row['id'] not in himgraph_2.nodes():
        himgraph_2.add_node(row['id'],name= row['name'],kind=row['kind'])
for index,row in dfhimedge_2.iterrows():
    himgraph_2.add_edge(row['source'],row['target'],metaedge = row['metaedge'])
```

```python
from node2vec_embed import node_2vec

model= node_2vec(test_graph)
```

```python
f1 = open('positive_list','r')
positive_list = simplejson.load(f1)
positive_list
```

```python
f2 = open('negative_list','r')
negative_list = simplejson.load(f2)
negative_list
```

```python
import node2vec
pair_list=positive_list
edge_vecs = []
edges_embs = node2vec.edges.HadamardEmbedder(keyed_vectors=model.wv)
for i in pair_list:
        node1=i[0]
        node2=i[1]
        vec= edges_embs[(node1,node2)].tolist()
        edge_vecs.append(vec)
print(len(edge_vecs))
```
